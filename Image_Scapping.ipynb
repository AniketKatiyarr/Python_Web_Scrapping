```python
query = "elon musk"
response = requests.get(f"https://www.google.com/search?q={query}&sca_esv=578202142&rlz=1C1CHBF_enIN975IN975&tbm=isch&sxsrf=AM9HkKkSfn2KITEKQEMIIfcKOfsLKgH8JA:1698772810494&source=lnms&sa=X&ved=2ahUKEwjDt7Hv5aCCAxUj2DgGHagUFgkQ_AUoAnoECAMQBA&biw=1536&bih=695&dpr=1.205")
```

```python
soup = BeautifulSoup(response.content,"html.parser")
```

```python
soup
```

```python
Image_Tages = soup.find_all("img")
Image_Tages
```

```python
del Image_Tages[0]
```

```python
len(Image_Tages)
```

```python
image_data_mongo = []
for i in Image_Tages:
    image_url = i['src']
    image_data = requests.get(image_url).content
    My_Dic = {"index" : image_url, "images": image_data}
    image_data_mongo.append(My_Dic)
    with open(os.path.join(save_dir,f"{query}_{Image_Tages.index(i)}.jpg"),"wb") as f:
        f.write(image_data)
          
    
```
